#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sinasc = pd.read_csv('https://raw.githubusercontent.com/vinisavordelli/exercicios_ebac/main/Ci%C3%AAncia%20de%20Dados/Exercicio_5/SINASC_RO_2019.csv', sep=',')
not_na = sinasc[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']].notna()
not_na = not_na.astype(int)
clean_df = pd.melt(not_na)
clean_df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, , ESTCIVMAE.

```python
fig, axes = plt.subplots(2, 3)
not_na = sinasc[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']].notna()
fig.set_size_inches(20, 10)


# Não entendi ao certo, está escrito que é para fazer um, mas gráfico está no plural. 
# Como não vi muito sentido em colocar todos em um gráfico só, assumi que era para fazer um para cada variável.
# Mas para colocar todos os gráficos de alguma maneira em um gráfico só, criei um gráfico que representa a quantidade de valores faltantes para cada variável.

sns.countplot(ax= axes[0, 0], x='APGAR1', data=sinasc)
sns.countplot(ax= axes[0, 1], x='APGAR5', data=sinasc)
sns.countplot(ax= axes[0, 2], x='SEXO', data=sinasc)
sns.countplot(ax= axes[1, 0], x='GESTACAO', data=sinasc)
sns.countplot(ax= axes[1, 1], x='ESTCIVMAE', data=sinasc)
na_values = sns.countplot(data=clean_df.loc[clean_df['value'] != True], x='variable', ax=axes[1, 2])
na_values.set_ylabel('Valores faltantes')
na_values.set_xticklabels(na_values.get_xticklabels(), rotation=45)




```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(y='CONSULTAS', x='APGAR5', data=sinasc, orient='h')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig2, axes2 = plt.subplots(2, 2)

sns.countplot(ax= axes2[0, 0], x='SEXO', data=sinasc)
sns.countplot(ax= axes2[0, 1], x='ESTCIVMAE', data=sinasc)
gestacao = sns.countplot(ax= axes2[1, 0], x='GESTACAO', data=sinasc)
sns.countplot(ax= axes2[1, 1], x='CONSULTAS', data=sinasc)

gestacao.set_xticklabels(gestacao.get_xticklabels(), rotation=45)
```
